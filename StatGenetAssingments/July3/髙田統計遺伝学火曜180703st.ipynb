```r
library(igraph)
set.seed(3)
M <- matrix(rnorm(100),10,10)
M <- (M + t(M))/2
M <- (M>-0.5)+0#true and false matrix
diag(M) <- 0
g <- graph.adjacency(M,mode="undirected")
E(g)$weight <- abs(round(log((rnorm(100)^2)*100)))
plot(g,edge.label=E(g)$weight)
```

```r
#ここからが最小全域木
g_mst <- mst(g)
plot(g_mst,edge.label=E(g)$weight)
```
