```r
#Make weighted complete graph
library(igraph)
ajmat <- matrix(rnorm(10*10), nrow=10,ncol=10)
ajmat <- ajmat + t(ajmat)
diag(ajmat) <- 0
g <- graph.adjacency(ajmat, weighted=TRUE,mode="undirected")
plot(g,directed=F)
```

```r
#Make minimum spanning tree
g_mst <- mst(g,weights=E(g)$weight)
plot(g_mst)
```
