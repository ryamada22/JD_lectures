# Make a graph that is connected and find its minimum spanning tree and draw both graphs.

```r
el<-matrix(c(1,2,1,4,2,3,2,5,3,6,4,5,4,7,5,6,5,8,6,9,6,10,7,8,8,9,9,10),byrow=TRUE,ncol=2)
el
```

```r
g.el<-graph.edgelist(el,directed=FALSE)
plot(g.el)
```

```r
E(g.el)$weight <- c(5,8,6,1,3,4,8,7,11,16,18,3,9,1)
plot(g.el,vertex.label=V(g.el)$name, edge.label=E(g.el)$weight, main = "weighted graph")
dev.off()
```

```r
mst<-mst(g.el)
mst
```

```r
plot(mst,edge.label=E(g.el)$weight)
```
