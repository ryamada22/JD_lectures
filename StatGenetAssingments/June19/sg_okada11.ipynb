```r
#Make an adjacency matrix for n vertices that have k separated components
#Input is vector of the number of vertices of k separated graph
sep_graph <- function(sep_members){
    n <- sum(sep_members) #n vertices
    k <- length(sep_members) #k separated grapf
    admat <- matrix(NA,n,n)
    idx <- 1
    for(i in 1:k){
        s <- sep_members[i]
        admat[idx:(idx+s-1),idx:(idx+s-1)] <- 1 #Each separated grapf is complete graph 
        if(i==k) break
        admat[idx:(idx+s-1),(idx+s):n] <- 0 #manipulating the rectangular blocks of "zero-throughout".
        admat[(idx+s):n,idx:(idx+s-1)] <- 0 #by manipulating the rectangular blocks of "zero-throughout".
        idx <- idx + s
    }
    diag(admat) <- 0
    return(admat)
}
```

```r
#Make adjuncy matrix, n=20,k=5
sep_members <- c(3,3,4,5,5)
admat1 <- sep_graph(sep_members)
```
