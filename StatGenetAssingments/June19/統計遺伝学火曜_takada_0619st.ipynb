```r
# Make an adjacency matrix for n vertices 
# that have k separated components, 
# by manipulating the rectangular blocks of "zero-throughout".
$$
n個の頂点を持つ隣接行列を作る
その頂点はk個の構成要素から成っている　→離れていると判断
０で構成された長方形のブロックを操作する方法でやりなさい。
$$
```

```r
#このブロックでnとkを操作します。
n <- 40
k <- 2
```

```r
#まずは対称行列を作成する
M <- matrix(runif(n*n,0,1),n,n)
M2 <- M + t(M)
M3 <- M2-1
diag(M3) <- 0
range(M3)
MX <- M3#調節用で残しておく

m3 <- mean(M3)
M4 <- M3 >= m3
image(M4)
M5 <- M4 +0
image(M5)
```

```r
j <- sample(n,k)#nからランダムにk個選ぶ
l <- length(j)
for ( i in 1:l ){
    M5[1:j[i]-1,j[i]:n] <- 0
    M5[j[i]:n,1:j[i]-1] <- 0
}
```

```r
image(M5)
```

```r
library(igraph)
g3 <- graph.adjacency(M5,mode="undirected")
plot(g3)
```
