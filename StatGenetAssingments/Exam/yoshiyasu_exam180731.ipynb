# グラフの応用について
グラフとは、一次元の単体複体のこと。グラフは位相空間としては非常に単純であるが、対象の相関や因果を記述する際には有用な道具である。対象を頂点、それらの関係を辺の有無で表し、必要なら距離や向きを付加することにより、対象の関係を量的に扱うことができる。最初から関係が明示されている場合だけでなく、対象の量的な性質からグラフを生成し、関係を見出すこともできる。また、そうしてできたグラフの性質をラプラシアンや全域木といった、より単純なものから調べることもできる。講義では応用の具体例を見ながら、こういったグラフの性質や構成について学んだ。これにより、グラフの図から分類や推定、考え方の大まかな方針がわかるようになった。このサーベイでは

・分子ネットワークのサブネットワークを取り出して、各々の役割を推定している。全体の構造でなく、末端の小さなグループが大事な役割を持っていると考えられている。

・薬とターゲットの関係をグラフ化する。グラフ化の際には、薬同士、ターゲット同士の類似性を用いて互いの近さや関係の有無を定める。近い薬は近いターゲットに働きかける、最短距離によって効率的な薬の候補を推定できる。

といった例が挙げられていた。手法の比較も行われていたが、グラフから取り出す情報量と分子ネットワーク／薬とターゲットの関係についての情報量の対応が実感できず、よくわからなかった。遺伝学の基礎的な内容について、学習方法を質問したり、自習時間を確保すべきだった。
