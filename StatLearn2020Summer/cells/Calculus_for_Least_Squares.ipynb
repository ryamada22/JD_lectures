# Calculus for Least Squares

Assume a relation formula below with its number of rows more than its number of columns,
$$
y \sim X\mathbf{a}
$$
なる関係式があり、$X$の行数が列数より多いとき、

The following should be minimzed, that is the method of least squares,
$$
\sum_{i=1}^n (y_i-\hat{y_i})^2 = (y-X\mathbf{a})^T \cdot (y-X\mathbf{a})
$$
を最小にするような$\hat{\mathbf{a}}$が、推定値であった。

そして

$\hat{\mathbf{a}}$, the minimizing estimates is calculated as below,

$$
\hat{\mathbf{a}} = (X^TX)^{-1}X^Ty
$$

で求まるのであった。

## The Simplest Case 最も単純な場合

$$
y \sim X\mathbf{a}
$$

When the number of samples is $n$ and the number of variables is $1$,

the above formula are

$$
\begin{pmatrix}y_1 \\ y_2 \\ ... \\ y_n \end{pmatrix} \sim \begin{pmatrix}x_1,x_2,...,x_n \end{pmatrix} a
$$

$$
\sum_{i=1}^n (y_i-\hat{y_i})^2 = \sum_{i=1}^n (y_i - a x_i)^2
$$

Defferentiation with $a$ is,

$$
\frac{d}{d a} \sum_{i=1}^n (y_i-\hat{y_i})^2 = \sum_{i=1}^n \frac{d}{d a} (y_i - a x_i)^2
$$
Further transformation :
$$
\frac{d}{d a} \sum_{i=1}^n (y_i-\hat{y_i})^2 = \sum_{i=1}^n -2 x_i ( y_i -a x_i) = 2 (a \sum_{i=1}^n x_i - \sum_{i=1} y_i)
$$

Therefore 

$\hat{a}$ that is the solution of $\frac{d}{d a} \sum_{i=1}^n (y_i-\hat{y_i})^2 =0$ is,

$$
\hat{a} = \frac{\sum_{i=1}^n y_i}{\sum_{i=1}^n x_i}
$$



## Partial Differential Equation 偏微分方程式

$$
f(\mathbf{a})=f((a_1,...,a_m))=\sum_{i=1}^n (y_i-\hat{y_i})^2 = (y-X\mathbf{a})^T \cdot (y-X\mathbf{a})
$$

The formula above is a quadratic equations of $m$ variables $(a_1,...,a_m)$, where $y$ and $X$ are given.

は、$y$,$X$が与えられているとき、$m$個の変数$(a_1,...,a_m)$の二次式となっているスカラー関数である。

The minimum of $$f(\mathbf{a})$$ shuold be given when the following is satisfied for all $a_i, i=1,2,...,m$.

今、この$f(\mathbf{a})$を最小にする$\mathbf{a}=(a_1,...,a_m)$とは、
$$
\frac{\partial f(\mathbf{a})}{\partial a_i}=0
$$
がすべての$a_i$について成り立つ場合である。


Using $z^Tw = w z^T$, that returns a scalar value, 

スカラーを返す$z^Tw$は$z^Tw = w z^T$であることを使うと

$$
f(\mathbf{a}) = (y-X\mathbf{a})^T \cdot (y-X\mathbf{a})
 = y^Ty -2(X\mathbf{a})^Ty+(X\mathbf{a})^T \cdot (X\mathbf{a})
$$

Further transformation gives, 
さらに変形して

$$
f(\mathbf{a}) = y^Ty -2(X^Ty)\cdot \mathbf{a} + \mathbf{a}^TX^TX\mathbf{a}
$$

Partial differentiation of the above with $\mathbf{a}$.

これを$\mathbf{a}$で偏微分する。

The 1st term is 0,

第1項は$0$

The 2nd term keeps each element of $-2y^TX$.

第2項は、各成分に$-2y^TX$の各成分が残る。

The 3rd term of partial derivative with $a_k$ becomes as below, using 

第３項は、$X^TX=Z$と置くと

$$
\mathbf{a}^T Z \mathbf{a} = \sum_{i=1}^m \sum_{j=1}^m a_i a_j Z_{ij}
$$

where $X^TX=Z$,

であり、その$a_k$による偏微分は

$$
\frac{\partial}{\partial a_k} (\sum_{i=1}^m \sum_{j=1}^m a_i a_j Z_{ij}) = 2\sum_{i=1}^m Z_{ik}a_i
$$

The sum over $k=1,...,m$ becomes,

となり、$k=1,...,m$について合わせると
$$
2Z \mathbf{a}
$$
となるから、

Then 
結局、

$$
0 - 2X^Ty + 2Z\mathbf{a} = 2(X^TX) \mathbf{a} - 2X^Ty = \mathbf{0}
$$

となり、

$$
(X^TX)\mathbf{a} = X^T y
$$
が得られる。


When $X^TX$ has its inverse,

ここから、$X^TX$に逆行列があるときは

$$
\mathbf{a} = (X^TX)^{-1}X^T y
$$

が得られる。
