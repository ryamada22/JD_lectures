# Bias and Variance バイアスとバリアンス

## Example 


```r
library(class)
K <- 200
p <- 2
x <- matrix(rnorm(K*p),ncol=p)

ctr1 <- 1
ctr2 <- 0.5
r1 <- 0.2
r2 <- 0.5
v <- r1 * (x[,1]-ctr1)^2 + r2 * (x[,2]-ctr2)^2
t <- 0.5
v0 <- which(v< t)
v1 <- which(v>=t)

gr <- rep(0,K)
p0 <- 0.1
p1 <- 0.9
gr[v0] <- sample(0:1,length(v0),replace=TRUE,prob=c(p0,1-p0))
gr[v1] <- sample(0:1,length(v1),replace=TRUE,prob=c(p1,1-p1))

plot(x,col=gr+1,pch=20,cex=2)

x1. <- seq(from=min(x[,1]),to=max(x[,1]),length=100)
x2. <- seq(from=min(x[,2]),to=max(x[,2]),length=100)
x. <- as.data.frame(expand.grid(x1.,x2.))
```

## When sample size is very big

```r
library(class)
K <- 20000
p <- 2
x <- matrix(rnorm(K*p),ncol=p)

ctr1 <- 1
ctr2 <- 0.5
r1 <- 0.2
r2 <- 0.5
v <- r1 * (x[,1]-ctr1)^2 + r2 * (x[,2]-ctr2)^2
t <- 0.5
v0 <- which(v< t)
v1 <- which(v>=t)

gr <- rep(0,K)
p0 <- 0.1
p1 <- 0.9
gr[v0] <- sample(0:1,length(v0),replace=TRUE,prob=c(p0,1-p0))
gr[v1] <- sample(0:1,length(v1),replace=TRUE,prob=c(p1,1-p1))

plot(x,col=gr+1,pch=20,cex=0.01)

x1. <- seq(from=min(x[,1]),to=max(x[,1]),length=100)
x2. <- seq(from=min(x[,2]),to=max(x[,2]),length=100)
x. <- as.data.frame(expand.grid(x1.,x2.))
```

```r
k <- 10
knn.out <- knn(x,x., gr, k = k, prob=TRUE)
plot(x.,col=as.numeric(knn.out)+2,pch=20)
points(x,col=gr+1,pch=20,cex=0.001)
```

Almost perfect boundary is estimated.

## The Truth

+ Standard normal distribution
$$
X \sim N(0,I)
$$
+ The space is divided into two regions.
$$
0.2(x_1-1)^ 2 + 0.5(x_2^2-0.5)^2 = 0.5
$$
+ The label values are stochastically selected; 0.1 vs. 0.9 in the ellipse and 0.9 vs. 0.1 out of the ellipse.


## The Best Prediction, But Still Uncertain

Assume you estimated the truth perfectly somehow.

Can you tell the label of the new sample after you know the location of the sample?

The answer is NO.

You can tell the region in which the sample exists, but you have to guess the label value (0.1 vs. 0.9) or (0.9 vs. 0.1), according to the region it belongs to.

This uncertainty, or the variance of values to be observed, exists because of the true process. Any estimation of the model can not overcome this uncertainty.

In other words, there is variation in values beyond the estimation.

## The Limit of Strategy of Estimation

+ When you attempt to estimate the TRUTH as above, the border between two regions, and the probability of label value in two regions, the variance of label values is due to the stochasticity within the truth.

+ You may only attempt to segregate the space into two types, one of which is "0-dominant area" and the other of which is "1-dominant area". This is the way the classification methods do. In this case, you would tell a new sample in "o-dominant area" should have label "0" with probability 1 and vice versa. 

+ There are further limits as follow.

## Limited number of samples.

As shown above, kNN method with very many samples identified the almost complete border curve for classification.

When the sample size is relatively small, the border curve is not good. See below.


```r
K <- 200
p <- 2
x <- matrix(rnorm(K*p),ncol=p)

ctr1 <- 1
ctr2 <- 0.5
r1 <- 0.2
r2 <- 0.5
v <- r1 * (x[,1]-ctr1)^2 + r2 * (x[,2]-ctr2)^2
t <- 0.5
v0 <- which(v< t)
v1 <- which(v>=t)

gr <- rep(0,K)
p0 <- 0.1
p1 <- 0.9
gr[v0] <- sample(0:1,length(v0),replace=TRUE,prob=c(p0,1-p0))
gr[v1] <- sample(0:1,length(v1),replace=TRUE,prob=c(p1,1-p1))

x1. <- seq(from=min(x[,1]),to=max(x[,1]),length=100)
x2. <- seq(from=min(x[,2]),to=max(x[,2]),length=100)
x. <- as.data.frame(expand.grid(x1.,x2.))

k <- 10
knn.out <- knn(x,x., gr, k = k, prob=TRUE)
plot(x.,col=as.numeric(knn.out)+2,pch=20)
points(x,col=gr+1,pch=20,cex=1)
```

Also the estimated border curve will change when another sample set is processed.

The border curves vary as shown.


```r
par(mfrow=c(3,3))
for(i in 1:9){
    x <- matrix(rnorm(K*p),ncol=p)

    ctr1 <- 1
    ctr2 <- 0.5
    r1 <- 0.2
    r2 <- 0.5
    v <- r1 * (x[,1]-ctr1)^2 + r2 * (x[,2]-ctr2)^2
    t <- 0.5
    v0 <- which(v< t)
    v1 <- which(v>=t)

    gr <- rep(0,K)
    p0 <- 0.1
    p1 <- 0.9
    gr[v0] <- sample(0:1,length(v0),replace=TRUE,prob=c(p0,1-p0))
    gr[v1] <- sample(0:1,length(v1),replace=TRUE,prob=c(p1,1-p1))

    x1. <- seq(from=min(x[,1]),to=max(x[,1]),length=100)
    x2. <- seq(from=min(x[,2]),to=max(x[,2]),length=100)
    x. <- as.data.frame(expand.grid(x1.,x2.))

    k <- 10
    knn.out <- knn(x,x., gr, k = k, prob=TRUE)
    plot(x.,col=as.numeric(knn.out)+2,pch=20)
    points(x,col=gr+1,pch=20,cex=0.1)
}
par(mfcol=c(1,1))
```

However when you repeat the procedures many times and average them, the average gets close to the ultimate goal.

The following panels show the result of averaging $$2^{0,1,...,8}$$ iterations.

The averaged results become smoother and seem to converge to somewhere.

```r
par(mfrow=c(3,3))
n.iter <- 2^(0:8)

cnt <- 0
for(i in 1:9){
    sum.knn.out <- rep(0,K)
    for(j in 1:n.iter[i]){
        x <- matrix(rnorm(K*p),ncol=p)

        ctr1 <- 1
        ctr2 <- 0.5
        r1 <- 0.2
        r2 <- 0.5
        v <- r1 * (x[,1]-ctr1)^2 + r2 * (x[,2]-ctr2)^2
        t <- 0.5
        v0 <- which(v< t)
        v1 <- which(v>=t)

        gr <- rep(0,K)
        p0 <- 0.1
        p1 <- 0.9
        gr[v0] <- sample(0:1,length(v0),replace=TRUE,prob=c(p0,1-p0))
        gr[v1] <- sample(0:1,length(v1),replace=TRUE,prob=c(p1,1-p1))

        x1. <- seq(from=min(x[,1]),to=max(x[,1]),length=100)
        x2. <- seq(from=min(x[,2]),to=max(x[,2]),length=100)
        x. <- as.data.frame(expand.grid(x1.,x2.))

        k <- 10
        knn.out <- knn(x,x., gr, k = k, prob=TRUE)
        sum.knn.out <- sum.knn.out + as.numeric(knn.out)
    }
    
    col <- as.numeric(sum.knn.out > (3*n.iter[i]/2))+2
    plot(x.,col=col,pch=20,main=paste("n=",n.iter[i]))
    points(x,col=gr+1,pch=20,cex=0.001)
}
par(mfcol=c(1,1))
```

## Converges to Wrong Answer : Bias

When we use kNN method with k=10 for sample sets with sample size 200, the average of results converges to the separating pattern similar to the n=256 panel above, which is clearly different from the TRUE separating curve.

In other words, kNN method with k=10 and sample ize =200 for this particular truth, THE ANSWER the method could achieve is not the TRUTH curve. 

This deviation of the ANSWER from the TRUTH is called "Bias".

## Results are deviated from the ANSWER : Variance

As shown above, individual separating curve based on one sample set is different from the ANSWER.

This deviation of each learing result from the ANSWER varies around "zero", and the variation is called "Variance".

## Randomness, Bias and Variance

There are 3 components that makes you to predict the next sample's label when you have preceding samples.

+ You get the separating curve based on a method and the samples. The result curve is deviated from the ANSWER. Variance.

+ The ANSWER is deviated from the TRUTH curve. Bias.

+ The label valus is stochastically determined. Therefore, the TRUTH curve is not enough to predict the label value. The INNATE UNCERTAINTY.
