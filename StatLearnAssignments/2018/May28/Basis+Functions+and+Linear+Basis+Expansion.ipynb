```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

### Draw the plane with $(\beta_1, \beta_2) = (2,-1)$ in 3d space

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.linspace(0, 10, 100)
rng = np.random.RandomState(0)
rng.shuffle(x1)
x1 = np.sort(x1[:20])
x2 = np.linspace(0, 10, 100)
rng = np.random.RandomState(1)
rng.shuffle(x2)
x2 = np.sort(x2[:20])
y = np.linspace(0, 10, 100)
rng = np.random.RandomState(2)
rng.shuffle(y)
y = np.sort(y[:20])

#x1 = x1.reshape((1, -1))
#x2 = x2.reshape((1, -1))

#x1, x2 = np.meshgrid (x1, x2)
b1 = 2
b2 = -1
y_plane = np.column_stack((b1*x1,b2*x2))
#y_plane = np.empty(shape=[x1, x2])
ax.plot_surface(x1, x2, y_plane.T)

ax.plot3D(x1, x2, y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.show()
```
