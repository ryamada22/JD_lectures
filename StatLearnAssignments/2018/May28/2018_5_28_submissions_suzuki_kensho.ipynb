$y = \beta_1 x_1 + \beta_2 x_2$ is a plane in 3d space where $x_1, x_2, y$ are three axes. Draw the plane with $(\beta_1, \beta_2) = (2,-1)$ in 3d space.

```python
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
x = np.arange(-3, 3, 0.5)
y = np.arange(-3, 3, 0.5)
X, Y = np.meshgrid(x, y)
```

```python
Z = 2 * X - 1 * Y

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter3D(np.ravel(X),np.ravel(Y),np.ravel(Z), )
```

$||y-X \beta ||^2$ is a function of $\beta$ when y and X are given. When $$ y = \begin{pmatrix} 1\\ 2\\ 3\\ 4 \end{pmatrix}$$ and $$X = \begin {pmatrix} 0,0\\ 1,0 \\ 0,1 \\ 1,1 \end{pmatrix},$$ draw a curved surface over grid points of $(\beta_1, \beta_2)$.

```python
fig2 = plt.figure()
ax2 = Axes3D(fig2)

y = np.matrix('1;2;3;4')
X = np.matrix([[0,0],[1,0],[0,1],[1,1]])

b_1 = np.arange(-100, 100, 0.5)
b_2 = np.arange(-100, 100, 0.5)
B_1, B_2 = np.meshgrid(b_1, b_2)

k = b_1.size
l = b_2.size

z_1 = np.zeros([k,l])
z_2 = np.zeros([k,l])
z_3 = np.zeros([k,l])
z_4 = np.zeros([k,l])

for i in range(k):
    for j in range(l):
        B = np.matrix([[B_1[i][j]],[B_2[i][j]]])
        Z = y - X * B
        z_1[i][j] = Z[0] ** 2
        z_2[i][j] = Z[1] ** 2
        z_3[i][j] = Z[2] ** 2
        z_4[i][j] = Z[3] ** 2

ax2.plot_wireframe(B_1,B_2,z_1,color='r',alpha=0.3)
ax2.plot_wireframe(B_1,B_2,z_2,color='g',alpha=0.3)
ax2.plot_wireframe(B_1,B_2,z_3,color='y',alpha=0.3)
ax2.plot_wireframe(B_1,B_2,z_4,color='b',alpha=0.3)
```

$y = \beta_1 x_1 + \beta_2 x_2$ can be written as $$y = \beta_1 h_1(x) + \beta_2 h_2(x),$$ where $(h_1(x),h_2(x)) = (x_1,x_2).$ It represented a plane in 3d space $(x_1,x_2,y)$ and a point in 2d space $(\beta_1,\beta_2)$. When $$(h_1(x),h_2(x)) = (x_1+x_2,x_1^2),$$$y = \beta_1 h_1(x) + \beta_2 h_2(x)$ is not a plane in 3d space but is a point in 2d space. Draw a curved surfacef in 3d space where $x_1, x_2, y$ are three axes.

```python
# suppose (b_1, b_2) = (2, -1)

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(x, y)

Z = 2 * (X + Y) - 1 * X ** 2

fig3 = plt.figure()
ax3 = Axes3D(fig3)

ax3.plot_surface(X,Y,Z,cmap=cm.coolwarm,alpha=0.5, linewidth=0, antialiased=False)
```
