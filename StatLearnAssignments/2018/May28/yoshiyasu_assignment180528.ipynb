・$y=\beta_1x_1+\beta_2x_2$ is a plane in 3d space where $x_1, x_2, y$ are three axes. Draw the plane with $(\beta_1, \beta_2) = (2,-1)$ in 3d space.

```r
x1 <- seq(-10, 10, length=100)
x2 <- x1
f <- function(x1, x2){
    h <- 2 * x1 - x2
}
y <- outer(x1, x2, f)
persp(x1, x2, y, theta=45, phi=45, expand=0.5)
```

・$||y-X \beta ||^2$ is a function of $\beta$ when y and X are given. When $$ y = \begin{pmatrix} 1\\ 2\\ 3\\ 4 \end{pmatrix}$$ and $$X = \begin {pmatrix} 0,0\\ 1,0 \\ 0,1 \\ 1,1 \end{pmatrix},$$ draw a curved surface over grid points of $(\beta_1, \beta_2)$.

$f(\beta_1,\beta_2):=\|y-X\beta\|$ と定める。上で与えられた $y, X$ を代入して、
$$
f(\beta_1,\beta_2)=\sqrt{(1-0)^2+(2-\beta_1)^2+(3-\beta_2)^2+(4-\beta_1-\beta_2)^2}=\sqrt{2(\beta_1^2+\beta_2^2+\beta_1\beta_2-6\beta_1-7\beta_2+15)}
$$
を描画すると以下のようになる。

```r
b1 <- seq(-10, 10, length=100)
b2 <- b1
f <- function(b1, b2){
    h <- (2 * (b1^2 + b2^2 + b1*b2 -6*b1 - 7*b2 + 15))^(1/2)
}
y <- outer(b1, b2, f)
persp(b1, b2, y, theta=45, phi=45, expand=0.5)
```

・$y = \beta_1 x_1 + \beta_2 x_2$ can be written as $$y = \beta_1 h_1(x) + \beta_2 h_2(x),$$ where $(h_1(x),h_2(x)) = (x_1,x_2).$ It represented a plane in 3d space $(x_1,x_2,y)$ and a point in 2d space $(\beta_1,\beta_2)$. When $$(h_1(x),h_2(x)) = (x_1+x_2,x_1^2),$$$y = \beta_1 h_1(x) + \beta_2 h_2(x)$ is not a plane in 3d space but is a point in 2d space. Draw a curved surface in 3d space where $x_1, x_2, y$ are three axes.

```r
x1 <- seq(-10, 10, length=100)
x2 <- x1
b <- matrix(rnorm(2),ncol=2)
b
f <- function(x1, x2){
    h <- b[1,1] * (x1 + x2) + b[1,2] * x1^2
}
y <- outer(x1, x2, f)
persp(x1, x2, y, theta=45, phi=45, expand=0.5)
```
