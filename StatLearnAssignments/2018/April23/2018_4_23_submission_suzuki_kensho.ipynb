### 機械学習におけるバイアス、バリアンス、およびこれらのトレードオフについて

教師あり学習における過学習・未学習の指標となるもので、これらの理論的な解析ができる。
数式的には、損失関数の期待値を分解することで、バリアンス、バイアス、ノイズを計算することができる。

- 未学習
    学習が不十分であるために、データを表現するモデル(関数)が最適化されていない状態。つまり、訓練データにおいて、モデルによる予測の正答率が十分に上昇していない(訓練誤差が減少していない)状態。
- 過学習
    未知のデータに対する予測性能が著しく低い状態。言い換えると、モデルの正答率を上げるために、訓練データのみを(無理に)表現するようなモデルになってしまた状態。


#### バイアス
学習がうまくいかない度合いのことで、モデルによる予測が真値からずれる度合いを示す。つまり、未学習の状態を表現する。一般的には、モデルが単純であるときにバイアスは大きくなる。

#### バリアンス
複数の訓練データに対して、モデルによる予測が訓練データ毎に変動する度合いのこと。もしくは、一つの訓練データに依存し過ぎることで、新しいデータへの予測性能が悪化する過学習の度合い。一般的にはモデルが複雑になるほどバリアンスも大きくなる。

#### トレードオフ
バイアスとバリアンスとの板挟みを表す言葉で、ある意味ジレンマである。バイアスを小さくしようとモデルを複雑にすると(もしくは正則化などによる過学習抑制を小さくすると)、バリアンスが大きくなり、逆にバリアンスを小さくしようとモデルを単純にすると、バイアスが大きくなる。損失関数の期待値が小さくなるようにすると、バイアス・バリアンス双方が小さい状態にすることができる。
