## データサイエンス手法のリスト

+ **maximum a posteriori (MAP)** 
　実測データから得られる事前分布に基づいて事後確率を最大にする推定方法である。条件付き確率を計算し、その推定モデルを基に解析を行うため統計解析に含まれると考える。推定の前に情報が与えられ、それは何らかの学習をするというようにも感じるが、それは推定を行う前にデータの傾向を用いて推定モデルを作ることであるため機会学習には該当しないと考える。


+ **k近傍法(KNN)**
　過去のデータでの似た者同士を学習させ、新規出現者がどちらにより似ているかを決定する。より学問的な表現を使用すると、学習データをベクトル空間上に表示し、新規のサンプルをその空間ベクトル上に置く。その新規サンプルから距離の近い順に任意のk個を取得し、その中で多い方のグループに新規サンプルを入れるという方法で分類する。分類のために使用されること、学習を基に新規サンプルに適用するという過程は機械学習であると考える。
 
 
+ **主成分分析**
　大量のデータにその大きな傾向を与えている因子が何かを見出す手法である。これはデータの分散が最大になるように次元を圧縮する方法で、教師なし機械学習の次元削除の手法とあった。参考資料では学習したルールを新規サンプルに適用するのが機械学習とあったため、この方法は機械学習に含まれないのではないかと考えた。しかしその教師がない機械学習とは傾向だけは与えるが、その傾向に寄与する因子(正解)は与えないというだけで、傾向は示され学習しており、その原因は何でもあり得るため、傾向を学習し正解を推定せよという方法も機械学習に含まれる。そのためこの分析も機械学習であると考えられる。


 機械学習を単に推定、予測前に何らかの情報が与えられることと考えていたがそうではない。
 K近傍法と主成分分析はいずれも機械学習に含まれるが前もって正解を与えられる教師ありと教師なしに分けられた。何れにしてもその傾向は学習でき、新規サンプルに適用するため機械学習と言える。しかしMAPは事前に学習ができるような情報が与えられているものの、推定モデルを作った上で新規サンプルに適用する、条件付き確率を用いていることから統計解析に含まれるのではないかと考える。
