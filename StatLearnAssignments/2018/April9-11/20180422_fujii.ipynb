+ How to evaluate learning results and learning methods

One of the purposes of machine learning is construct the model to understand the given dataset and predict the unknown dataset.

There are at least two points to be evaluated. One is internal validity that explains the accuracy of the model of the given dataset. The other is external validity that generalizes the model to any dataset.

+ <a href="https://en.wikipedia.org/wiki/Training,_test,_and_validation_sets">Training set and test set</a>

Training dataset is used to construct the model. Test dataset is used to evaluate the validity of the consructed model.

+ Bias and variance

<a href="https://en.wikipedia.org/wiki/Bias_of_an_estimator">Bias</a> is an error of the model and training set.

<a href="https://en.wikipedia.org/wiki/Variance">Variance</a> is an error of the model and test set


+ <a href="https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff">Bias-Variance tradeoff</a>

In linear algebra, varianve $V(f)$ is decomposed as

$$V(f)=E(f^2)-E(f)^2$$

$$\iff E(f^2)=E(f)^2+V(f)$$.

Bias and variane coorespond to $E(f)^2$ and $V(f)$, respectively.

The best way is to find the acoordance the sum of bias and variance but these two relate with each other.

+ Overfitting

If the model has many or the equal number of parameters to the samples (size of training dataset), the model can express the dataset almost completely. However, this model cannot be generalized because this model is too much fiited to training set. This model is not useful other than training set.


+ Model complexity

The more the model has parameters, the more it gets to be complex.
