# 1. データサイエンス手法のリスト
## 1. 過去に使った手法
- 統計学的手法 
  - 検定および推定 (連続データおよび離散データ)
  - 回帰分析
    - 生存時間解析 (Cox proportional hazards model)
    - Logistic回帰モデル
  - 傾向スコア法
  
2016年より医療の支払いデータを利用した臨床疫学研究を行なっています。主には生存時間解析を行うことが多く、現在勉強中です。普段はSASを利用していますが、昨年からRも使い始めました。
