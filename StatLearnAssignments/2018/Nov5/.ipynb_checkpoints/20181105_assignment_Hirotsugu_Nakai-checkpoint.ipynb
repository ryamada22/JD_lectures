dirichlet distributionを使用

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
Nh = 3000
Nd = 1000
N = Nh + Nd
```

```python
# dirichlet distributionを使用し、Xの状況8分類に関して、Y=1となる確率を作成
np.random.seed(1)
prob_y1_on_x = np.random.dirichlet(np.ones(8),size=1)[0] #0をいれないと2次元のnparrayになる
plt.plot(prob_y1_on_x,marker="o")
```

```python
# Xの状況8分類において、yが1になる相対割合（検査のcostがあがるにつれて上がるべき）
# Xの数字があがるにつれて、costのかかる検査になることを想定
y1_ratio_to_y0plus1 = np.arange(0.1,0.9,0.1)
plt.plot(y1_ratio_to_y0plus1,marker="o")
```

```python
# Xの状況8分類において、yが0になる相対割合　（足しても1にはならない）
y0_on_x = prob_y1_on_x/y1_ratio_to_y0plus1 - prob_y1_on_x
y0_on_x
```

```python
# Xの状況8分類において、yが0になる確率　（上記を確率に変更）
prob_y0_on_x = (y0_on_x/np.sum(y0_on_x))
prob_y0_on_x
```

```python
#作成した確率分布 (X3〜X1の順に、index0:000、index1:001,index2:010,,,,index7:111)
pd.DataFrame({"pro_y0":prob_y0_on_x,"pro_y1":prob_y1_on_x})
```

```python
# 検査のcostに応じた、accuracyの上昇もうまくいっていそう
test = prob_y1_on_x/prob_y0_on_x
import matplotlib.pyplot as plt
plt.plot(test)
plt.show()
```

```python
# 作成した確率分布prob_y0_on_xに基づき、y=0のデータ個数を設定　（順に000,001,010,011）
np.random.seed(3)
Y0cnt = np.random.multinomial(Nh,prob_y0_on_x)
Y0cnt
```

```python
# 確率分布prob_y0_on_xに基づくy=1のデータ個数
np.random.seed(4)
Y1cnt = np.random.multinomial(Nd,prob_y1_on_x)
Y1cnt
```

```python
#y=0およびy=1のデータ個数を横に連結
num = np.hstack((Y0cnt,Y1cnt))
num
```

```python
# 参考ページ　https://qiita.com/junkls/items/10384950963056cc8e08
# itertoolsにはpermutation,combination,product等があるが、今回はproductを採用

import itertools
comb =[]
l =[0,1]
for i in itertools.product(l, repeat=4):
    comb.append(np.array(i))
comb = np.array(comb)
print(comb)
```

```python
data = np.array([]).reshape(-1,4).astype("int64") #data.shape:(0,4)
for i in range(len(num)):
    d = np.tile(comb[i],num[i]).reshape(-1,4)
    data = np.vstack((data,d))
print(data) #nd arrayでshape(4000,4)のデータ
```

```python
#見やすいようにDataFrameに変えておく
import pandas as pd
df = pd.DataFrame(data=data,columns=['y', 'x3', 'x2','x1'])
df
```

```python
# iloc: 行や列の位置に基づいて行・列を取得
X = df.iloc[:,1:]
X.head(5)
```

```python
y = df.iloc[:,:1]
y.head(5)
```

```python
# decision treeの作成

from sklearn import tree
clf = tree.DecisionTreeClassifier().fit(X,y)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph
```
