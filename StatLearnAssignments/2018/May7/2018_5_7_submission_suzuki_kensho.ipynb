#### The last step of stepwise method returns the same result with LASSO and Ridge with their regularization parameter $\lambda = 0$. How do the plots show this phenomenon? This "goal" is "Full Model".

$\lambda = 0$であるならば、stepwise, LASSO, Ridge回帰は全て、最小二乗法と等価になる。
上のグラフにおいて、各横軸は$\lambda$が10000から始まり、0で終了している。従って、$\lambda = 0$の時は、各パラメータの取る値(縦軸)はそれぞれのグラフの右端であり、その値はこれら3つのグラフにおいて全て同じ値となっていることが確認できる。

#### How do the plots show "subset selection" of stepwise and LASSO methods ?

$\lambda$が大きくなる(横軸において左方向に進む)につれて、0となるパラメータが増えていく。このパラメータが0になることは、対応する変数の係数が0になることであり、回帰に使用する変数の選択がなされているとみなすことが出来る。


#### How do the plots show "no subset selection" of Ridge method?

$\lambda$が大きくなろうとも、0となるパラメータが無い。つまり、回帰において係数が0になる変数がなく、使用される変数の選択が起こらないことを意味する。ただし、$\lambda$が大きくなり過ぎると、パラメータの値は極小になる。
