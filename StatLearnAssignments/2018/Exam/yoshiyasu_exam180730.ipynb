# 機械学習で出来ることについて
機械学習はアルゴリズムであり、自動的に全てを整えてくれるわけではない。予め決めたアルゴリズムに基づき、膨大なデータをきちんと解析してくれるものである。この理解から、データセットの前処理の必要性、データセットのシャッフルや分割といった各処理の意味について読み解くことができた。Tip 2 で触れられていた三つのグループへの分割は知らなかったが、二つのグループへの分割について十分扱っていたため、ハイパーパラメータを設定するためのグループが独立に必要であることも容易に理解できた。

# アルゴリズムの選定について
無闇に複雑なアルゴリズムを選べば良いというものではない。それぞれの長短・向き不向きを把握した上で、問題に合わせてアルゴリズムを選ぶ必要がある。この理解から、自分の問題をアルゴリズミックに分解することの有用性、選定に迷ったら仕組みが単純なものから試していく、といった方針に納得することができた。また、kNN や k-means clustering といった具体例を扱っていたため、期待する特徴量がある場合と無い場合におけるアルゴリズムの代表例についても納得することができた。

# オーバーフィッティングと対処について
過学習という単語は、受講前からよく耳にしていた。将棋・囲碁の AI の敗因を説明する際によく使用されていたように記憶している。トレーニングデータセットへの対処だけに特化して、アルゴリズムが複雑化されることによる影響であることは早々に理解できた。オーバーフィッティングへの対処方法、正則化の具体例を数式抜きでグラフから読み取ることは難しかったが、コンセプトをつかむための読み方が少しできるようになった。Tip 7 の内容は概ねこの中に含まれていたため、悩まず読むことができた。
