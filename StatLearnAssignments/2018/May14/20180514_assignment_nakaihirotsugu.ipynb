## Also write the formula for three methods and add description on them.

## 1 Stepwise
一つずつ説明変数を追加したり削除したりしながら、
$$\begin{aligned}
\boldsymbol{S}_{\lambda}(\boldsymbol{\beta})  = ||\boldsymbol{y}-\boldsymbol{X\beta}||^2
\end{aligned}$$


が最小となるように最適な説明変数の組み合わせを選択。<br>


## 2 Lasso (Least absolute shrinkage and selection operator) 
$$\begin{aligned}
\boldsymbol{S}_{\lambda}(\boldsymbol{\beta})  = ||\boldsymbol{y}-\boldsymbol{X\beta}||^2 + \lambda||\boldsymbol{\beta}||_1 
\end{aligned}$$
最小二乗法の数式に$\lambda||\boldsymbol{\beta}||_1 $という正則化（罰則）項を加えたもの。
λ=0の時はStepwise法と同様だが、λ＞0の時、$ ||\hatβ|| $は小さくなりsubset selectionが行われる。
stepwise法と比べ、λに対するβの推定値の推移は小さくなっている。（Regularization）


## 3 Ridge: L2-regularized least square minimization
$$\begin{aligned}
\boldsymbol{S}_{\lambda}(\boldsymbol{\beta})  = ||\boldsymbol{y}-\boldsymbol{X\beta}||^2 + \lambda||\boldsymbol{\beta}||^2
\end{aligned}$$
これは上記のLassoの正則化項を二次にしたもので、L2正則化と言われる。
subset selectionはなく全ての変数が使われることになる。
$ ||\hat{β}|| $の値に制限(Regularization)がかかり、λに対するβの推移は小さくなっている。
