```python
from numpy.random import *
import statistics
import math
import matplotlib.pyplot as plt

N = 1000000
seed(100)  
rnd_N = randn(N)

##Generate the plot of normal distribution
print(statistics.mean(rnd_N))
plt.hist(rnd_N, bins=100)
plt.show()
```
