# Joint Probability

N=4000

### Target variable
Y::= 0 | 1

### Clinical variables
X1::= 0 | 1

X2::= 0 | 1

X3::= 0 | 1


```python
N=4000


#各分岐の確率の決定
ratio_of_bifircation={}

##辞書keyは， Y X1 X2 X3 の値を順に並べたものであり，tree上のnodeの位置を指定している。 
##辞書valueは，辞書keyにより指定されたnodeの位置への，一つ前のnodeからの分岐の割合(確率)を示しており，2個1組の合計値は1である( p + (1-p) = 1 )。

### root
ratio_of_bifircation['0']=3000/4000
ratio_of_bifircation['1']=1000/4000

### root→Y=0
p=2/3
ratio_of_bifircation['00']=p
ratio_of_bifircation['01']=1-p

### root→Y=1
p=1/5
ratio_of_bifircation['10']=p
ratio_of_bifircation['11']=1-p

### root→Y=0 → X1=0
p=9/10
ratio_of_bifircation['000']=p
ratio_of_bifircation['001']=1-p

### root→Y=0 → X1=1
p=7/10
ratio_of_bifircation['010']=p
ratio_of_bifircation['011']=1-p

### root→Y=1 → X1=0
p=1/2
ratio_of_bifircation['100']=p
ratio_of_bifircation['101']=1-p

### root→Y=1 → X1=1
p=5/8
ratio_of_bifircation['110']=p
ratio_of_bifircation['111']=1-p

### root→Y=0 → X1=0 → X2=0
p=17/18
ratio_of_bifircation['0000']=p
ratio_of_bifircation['0001']=1-p

### root→Y=0 → X1=0 → X2=1
p=4/5
ratio_of_bifircation['0010']=p
ratio_of_bifircation['0011']=1-p

### root→Y=0 → X1=1 → X2=0
p=5/7
ratio_of_bifircation['0100']=p
ratio_of_bifircation['0101']=1-p

### root→Y=0 → X1=1 → X2=1
p=14/15
ratio_of_bifircation['0110']=p
ratio_of_bifircation['0111']=1-p

### root→Y=1 → X1=0 → X2=0
p=4/5
ratio_of_bifircation['1000']=p
ratio_of_bifircation['1001']=1-p

### root→Y=1 → X1=0 → X2=1
p=9/10
ratio_of_bifircation['1010']=p
ratio_of_bifircation['1011']=1-p

### root→Y=1 → X1=1 → X2=0
p=47/50
ratio_of_bifircation['1100']=p
ratio_of_bifircation['1101']=1-p

### root→Y=1 → X1=1 → X2=1
p=0
ratio_of_bifircation['1110']=p
ratio_of_bifircation['1111']=1-p

print(ratio_of_bifircation)
```

```python
#サンプルの作成
rb=ratio_of_bifircation
samples=[]

for Y in range(2):
    for X1 in range(2):
        for X2 in range(2):
            for X3 in range(2):
                
                noden=str(Y)+str(X1)+str(X2)+str(X3) #nodeの位置を示す番号。辞書keyに使用。
                print("Node No. ="+noden )
                
                Pr=rb[noden[0:1]]*rb[noden[0:2]]*rb[noden[0:3]]*rb[noden[0:4]] #各leafに含まれるサンプルの割合(確率)
                print(" {0}".format(Pr*N)) ##各leafに含まれるサンプル数
                
                for i in range(round(Pr*N)):
                    samples.append([Y, X1, X2, X3])
print('\n')
print(samples) #[Y, X1, X2, X3]
                
```
