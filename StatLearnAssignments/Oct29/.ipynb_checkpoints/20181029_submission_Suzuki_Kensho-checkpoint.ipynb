```python
import random
import numpy as np
from graphviz import Digraph, Source
```

```python
dot = Digraph(name="bifurcation_tree", filename="bifurcation_tree", format='png')
```

```python
node_values = {0: 4000}
fraction_vectors = [3/4, 2/3, 1/5, 3/5, 3/5, 1/4, 3/8, 1/3, 1/4, 1/6, 1/8, 9/10, 2/3, 1/3, 4/5]
node_shapes = ["ellipse", "diamond", "circle", "box", "doublecircle"]
node_colors = ["black", "yellow", "red", "green", "blue"]
label_names = ["root", "Y", "X1", "X2", "X3"]
frac_prod = [1 for i in range(2**5-1)]
```

```python
for i in range(5):
#     dot.attr(label=label_name[i], fontcolor="black")
    dot.attr("node", shape=node_shapes[i], color=node_colors[i])
    for j in range(2**i):
        num = 2**i+j-1
        n = num-1
        if i == 0:
            dot.node(str(i), "total\n{}\n{}".format(str(frac_prod[num]*100)+"%", node_values[num]))
        else:
            pre_node = n//2
            pos_or_neg = 0 if j%2 == 0 else 1
            frac = fraction_vectors[pre_node] if j%2 == 0 else (1-fraction_vectors[pre_node])
            node_values[num] = round(node_values[pre_node] * frac)
            for k in range(i):
                frac_prod[num] *= (fraction_vectors[pre_node] if n%2 == 0 else (1-fraction_vectors[pre_node]))
                n = (n-2)//2
                pre_node = n//2
            label = "{} : {}\n{}\n{}".format(label_names[i], pos_or_neg, str(round(frac_prod[num]*100, 3))+"%", node_values[num])
            dot.node(str(num), label)

for i in range(2**(5-1)-1):
    dot.edge(str(i), str(2*i+1), label=str(round(fraction_vectors[i],2)))
    dot.edge(str(i), str(2*i+2), label=str(round(1-fraction_vectors[i],2)))
```

```python
Source(dot.source)
```

```python
dot.render(cleanup=True)
```
