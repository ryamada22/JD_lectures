# 20181015 Dicision Trees Classnote
* non-parametric supervised learning method for regression and classification.
* The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.
* The deeper the tree, the more complex the decision rules and the fitter the model.
* 変数(discreat)が1つしかなければ、「記述統計」しかできない

* all nodes are connected, a-cyclic が treeの定義
* input data X はどんな変数でも良い(categoricalでもcontinuousでも、orderedでも大丈夫)
* 通常outputは2個(dechotomous categoryという)で inputはmultipleなcontinuousなvalueであることが多い simpleなルールであった方が良い
* [こちら](http://scikit-learn.org/stable/modules/tree.html)

* x1< a → red or blueなどそういう条件を狭めていくと、基本的には長方形で空間を切っていくことになる、それがdesicion treeのパターン
* つまり、条件式が複雑だと計算が大変なので x1 < a など不等号で片方がopenな状態でやることが多い
* あちこちに色々分布しているもの、赤、青、赤、と分けたいときとか、オーダーとか関係なく、分けたい時とか。つまり<mark>長方形で区切りやすい分布には有用<mark>

[こちら](http://scikit-learn.org/stable/modules/tree.html)　画像認識などはdicision treesが有用かも、、

```python
from sklearn import tree
X = [[0, 0], [1, 1]]#Y(標本)2つあり、それぞれの標本の説明変数が2つずつある
Y = [0, 1]#標本が2つある
X
Y
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
#[2.,2.]というinputがどのようなYになるか予測 1だよという結果
clf.predict([[2., 2.]])
```

```python
#[2.,2.]というinputについてYが0,1になるそれぞれの確率を求めている
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
clf = tree.DecisionTreeClassifier()
clf = clf.fit(iris.data, iris.target)
```

```python
#decision treeをグラフ化する
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("iris") 
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
